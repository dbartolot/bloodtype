## define model


```python
model = BloodType(10000, timesteptype='m')
```

```python
t = trange(5000)
for i in t:
    t.set_description("Population Size {}".format(model.populationsize))
    t.refresh()
    model.step()
model.print_state()
```

# Model with Rhesus Factor

```python
model2 = BloodType(1000, timesteptype='m', birthRate=0.13, deathRate=0.115)

# iterations

for i in tqdm(range(500)):
    model2.step(rf_mutation='-', mutations=10)
model2.printState()


for i in tqdm(range(2000)):
    model2.step()
model2.printState()


model2.plotBtPt(showRf=True)
```
